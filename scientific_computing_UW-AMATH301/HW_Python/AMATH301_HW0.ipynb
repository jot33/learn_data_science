```python
# Problem 2

# Define domain
dx = 0.1
x = np.arange(-10,10+dx,dx)

# Define Function
f = lambda x: x*np.sin(3*x) - np.exp(x)
flabel = 'f(x) = xsin(3x) - exp(x)'

# Define 1st Derivative of Function
df = lambda x: 3*x*np.cos(3*x) - np.exp(x) + np.sin(3*x)
dflabel = 'df/dx = 3*x*cos(3*x) - exp(x) + sin(3*x)'

# Plot function to verify
plt.style.use('dark_background') # Comment out for light background
plt.figure(figsize=(15,8))
plt.plot(x,f(x),label=flabel)
plt.plot(x,df(x),label=dflabel)
# plt.xlim(x[0],x[len(x)-1])
# plt.ylim(-15,15)
plt.axis([x[0],x[len(x)-1],-15,15])
plt.legend();
```

```python
def nrsolv(fc,dfc,x0,tol):
    """
    Newton-Raphson Method
    x0 = initial guess
    fc = defined function
    dfc = 1st derivative of defined function
    tol = tolerance
    """
    # Run max 100 iterations
    for j in range(1, 101):
        if ( abs(fc(x0)) < 10e-6 ):
            display("Convergence at x = " + str(x0))
            display("Iterations: " + str(j))
            break
        elif j == 100:
            display("No convergence!")
        else:
            x0 = x0 - fc(x0)/dfc(x0)
```

```python
nrsolv(f,df,1,10e-6)
```
